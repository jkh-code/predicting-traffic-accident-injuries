# Presentation Graphs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("ggplot")
%matplotlib inline
```

```python
d = {
    "training_data": ["Injury/Non-injury data, no extra columns", "Injury/Non-injury data, extra columns", "Injury-only data, no extra columns", "Injury-only data, extra columns"],
    "Dummy Model": [0.5560, 0.5473, 0.8363, 0.8243],
    "Linear Regression": [0.5153, 0.5153, 0.7884, 0.7890],
    "Random Forest Reg.": [0.5267, 0.5270, 0.8077, 0.8056],
    "Gradient Boosted Reg.": [0.5107, 0.5111, 0.7842, 0.7852]}
```

```python
df = pd.DataFrame.from_dict(d)
df = df.set_index("training_data")
df
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
df.plot(kind="line", ax=ax)
ax.set_xticks(range(0, 3+1))
ax.set_xticklabels(["Injury/Non-injury data\nno extra columns", "Injury/Non-injury data\nextra columns", "Injury-only data\nno extra columns", "Injury-only data\nextra columns"])
ax.set_yticks(np.linspace(0.50, 0.85, 8))
ax.set_yticklabels([f"{num:0.2f}" for num in np.linspace(0.50, 0.85, 8)]);
ax.set_title("RMSE's For Given Dataset")
ax.set_xlabel("Dataset")
ax.set_ylabel("RMSE")
fig.tight_layout()
plt.savefig("./../images/rmses-for-given-dataset.png");
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
df.plot(kind="bar", ax=ax)
ax.set_xticks(range(0, 3+1))
ax.set_xticklabels(["Injury/Non-injury data\nno extra columns", "Injury/Non-injury data\nextra columns", "Injury-only data\nno extra columns", "Injury-only data\nextra columns"], rotation=0)
ax.set_yticks(np.linspace(0.0, 0.8, 9))
ax.set_yticklabels([f"{num:0.2f}" for num in np.linspace(0.0, 0.8, 9)]);
ax.set_title("RMSE's For Given Dataset")
ax.set_xlabel("Dataset")
ax.set_ylabel("RMSE")
fig.tight_layout()
plt.savefig("./../images/rmses-for-given-dataset-bar.png");
```
